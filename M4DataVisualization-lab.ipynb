### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Establish a connection to your SQLite database
conn = sqlite3.connect('m4_survey_data.sqlite')  # Ensure the correct database file name

# Query to select all data from the 'master' table
query = "SELECT * FROM master"

# Load data into a DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Filter the data for the age group 30 to 35
age_filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median of 'WorkWeekHrs' and 'CodeRevHrs'
median_workweekhrs = age_filtered_df['WorkWeekHrs'].median()
median_coderevhrs = age_filtered_df['CodeRevHrs'].median()

# Create a DataFrame for plotting
median_df = pd.DataFrame({
    'Category': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median Hours': [median_workweekhrs, median_coderevhrs]
})

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(median_df['Category'], median_df['Median Hours'], color=['skyblue', 'lightgreen'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Category')
plt.ylabel('Median Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Establish a connection to your SQLite database
conn = sqlite3.connect('m4_survey_data.sqlite')  # Ensure the correct database file name

# Query to select all data from the 'master' table
query = "SELECT * FROM master"

# Load data into a DataFrame
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Assuming df is your DataFrame containing 'Age' and 'ConvertedComp' columns
# Filter the DataFrame for ages between 45 and 60
age_filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate the median ConvertedComp for each age
median_comp_by_age = age_filtered_df.groupby('Age')['ConvertedComp'].median()

# Plot the data using a line chart
plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='b')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xticks(median_comp_by_age.index)  # Ensure each age is labeled on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
